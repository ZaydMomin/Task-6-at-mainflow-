```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.hist(df['Age'], bins=20, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Distribution of Age among Diabetes Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your DataFrame is named 'df' and the gender column is named 'Gen'
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.title('Gender Distribution among Diabetes Patients')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.xticks(gender_counts.index, ['Female', 'Male'])  # Changed labels to 'Female' and 'Male' for 0 and 1 respectively
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Diagnosis', y='BMI', data=df)
plt.title('Relationship between BMI and Diabetes Diagnosis')
plt.xlabel('Diagnosis (0: No Diabetes, 1: Diabetes)')
plt.ylabel('BMI')
plt.show()
```

```python
# Create BMI categories (you might need to adjust these based on your data)
bmi_bins = [0, 18.5, 25, 30, np.inf]  # Underweight, Normal, Overweight, Obese
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['BMI_Category'] = pd.cut(df['BMI'], bins=bmi_bins, labels=bmi_labels)

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.countplot(x='BMI_Category', hue='Diagnosis', data=df)
plt.title('Relationship between BMI and Diabetes Diagnosis')
plt.xlabel('BMI Category')
plt.ylabel('Number of Patients')
plt.legend(title='Diagnosis', labels=['No Diabetes', 'Diabetes'])
plt.show()
```

```python
# Assuming DataFrame is named 'df', smoking status column is 'Smoking', and diagnosis column is 'Diagnosis'

# Create a cross-tabulation (contingency table)
smoking_diabetes_ct = pd.crosstab(df['Diagnosis'], df['Smoking'])

# Calculate proportions
smoking_diabetes_props = smoking_diabetes_ct.div(smoking_diabetes_ct.sum(axis=1), axis=0)

# Create the stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))
smoking_diabetes_props.plot(kind='bar', stacked=True, ax=ax, color=['steelblue', 'red'])
plt.title('Relationship between Smoking Status and Diabetes')
plt.xlabel('Diagnosis (0: No Diabetes, 1: Diabetes)')
plt.ylabel('Proportion')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.legend(title='Smoking Status', labels=['Non-Smoker', 'Smoker'])
plt.show()
```

```python
# Assuming DataFrame is named 'df' and has columns 'FamilyHistoryDiabetes' and 'Diagnosis'

# Create a contingency table
family_history_diabetes_ct = pd.crosstab(df['Diagnosis'], df['FamilyHistoryDiabetes'])

# Create a stacked bar chart to show proportions
family_history_diabetes_props = family_history_diabetes_ct.div(family_history_diabetes_ct.sum(axis=1), axis=0)
fig, ax = plt.subplots(figsize=(8, 6))
family_history_diabetes_props.plot(kind='bar', stacked=True, ax=ax, color=['linen', 'sienna'])
plt.title('Relationship between Family History and Diabetes Diagnosis')
plt.xlabel('Diagnosis (0: No Diabetes, 1: Diabetes)')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.legend(title='Family History', labels=['No Family History', 'Family History'])
plt.show()
```

```python
# 6. Correlation between Socioeconomic Status and Diabetes Diagnosis (Stacked Bar Chart)
# Assuming you have an 'SocioeconomicStatus' column in your DataFrame

ses_diagnosis_ct = pd.crosstab(df['SocioeconomicStatus'], df['Diagnosis'])
ses_diagnosis_props = ses_diagnosis_ct.div(ses_diagnosis_ct.sum(axis=1), axis=0)
fig, ax = plt.subplots(figsize=(10, 6))
ses_diagnosis_props.plot(kind='bar', stacked=True, ax=ax, color=['skyblue', 'lightcoral'])
plt.title(' Correlation between Socioeconomic Status and Diabetes Diagnosis')
plt.xlabel('SocioeconomicStatus')
plt.ylabel('diagnosis')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.legend(title='Diagnosis', labels=['No Diabetes', 'Diabetes'])
plt.show()
```

```python
# Assuming your DataFrame is named 'df'
# Filter data for patients diagnosed with diabetes (Diagnosis == 1)
diabetes_patients = df[df['Diagnosis'] == 1]

# Calculate the number of patients using and not using antidiabetic medications
medication_counts = diabetes_patients['AntidiabeticMedications'].value_counts()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(medication_counts, labels=['Not Using', 'Using'], autopct='%1.1f%%', startangle=90, colors=['purple', 'yellow'])
plt.title('Proportion of Diabetes Patients Using Antidiabetic Medications')
plt.show();
```
